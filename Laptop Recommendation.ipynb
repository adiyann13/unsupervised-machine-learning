```python
## conecrting the non numeric to numeric values
from sklearn.feature_extraction.text import CountVectorizer
count_vectr = CountVectorizer(stop_words='english',max_features=6000 , binary=True)
vectorise = count_vectr.fit_transform(laptop_data_recommendation['tags']).toarray()
vectorise
```

```python
#### finding the similarity using cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similairty = cosine_similarity(vectorise)
similairty
```

```python
##buildin recommendation system for laptops
def recomm_laptop_name(laptops):
    indxs = laptop_data_recommendation[laptop_data_recommendation['laptop_name'] == laptops].index[0]
    dist =  sorted(list(enumerate(similairty[indxs])), reverse=True, key=lambda x: x[1])
    for i in dist[1:6]:
        print(laptop_data_recommendation.iloc[i[0]].laptop_name)
    
```
