```python
#seperating the the object type and numeric type for imputation
credit_data_obj = credit_data.select_dtypes(include = "object")
credit_data_obj

credit_data_numeric = credit_data.select_dtypes(exclude = "object")
credit_data_numeric
```

```python
##
from sklearn.impute import SimpleImputer
credit_data_numeric_impute = SimpleImputer(strategy = "median").fit_transform(credit_data_numeric)
credit_data_numeric_impute = pd.DataFrame(credit_data_numeric_impute , columns = credit_data_numeric.columns )
credit_data_numeric_impute
```

```python
credit_data_numeric_impute.isnull().sum()
```

```python
# merging both the dataframes now
credit_Data_final = pd.concat([credit_data_obj,credit_data_numeric_impute],axis=1)
credit_Data_final
```

```python
# dropping the cust id column
credit_Data_final.drop("CUST_ID" ,axis=1 ,inplace=True)
credit_Data_final
```

```python
##feature scaling
from sklearn.preprocessing import StandardScaler
credit_data_scaled = StandardScaler().fit_transform(credit_Data_final)
credit_data_scaled = pd.DataFrame(credit_data_scaled , columns = credit_Data_final.columns )
credit_data_scaled
```

```python
#lets do the clusteringusing aggleromwerative method
from scipy.cluster.hierarchy import linkage, dendrogram
import scipy.cluster.hierarchy  as sch

plt.figure(figsize=(15,14))
z_linkage = linkage(credit_data_scaled,method='ward')
z_linkage
dendro = dendrogram(z_linkage , color_threshold=10,truncate_mode='lastp',p=60)
dendro
plt.show()
```

```python
cluster_18 = sch.fcluster(z_linkage,18,criterion='maxclust')
cluster_18.tolist()
```

```python
credit_Data_final["clusters18_dendero"] = cluster_18
```

```python
credit_Data_final
```

```python
## doing the clistering using kmeans
from sklearn.cluster import KMeans
kmean_clustering = KMeans(n_clusters=18,random_state=47).fit(credit_data_scaled)
kmean_clustering
```

```python
kmean_clustering.labels_.tolist()
```

```python
credit_Data_final["k_means_clusters18"] = kmean_clustering.labels_
credit_Data_final
```
